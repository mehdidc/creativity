```python
%aimport lasagne.datasets.infinite_image_dataset
from lasagne.datasets.infinite_image_dataset import trans_scale_rotate_f, InfiniteImageDataset
import numpy as np
from skimage.data import lena, camera

import lasagne.easy

from lasagne.easy import get_2d_square_image_view, linearize
from pylearn2.scripts.plot_weights import grid_plot


import matplotlib.pyplot as plt
%matplotlib inline
```

```python

X = camera()[np.newaxis, :] / 255.

rgb = True if len(X.shape)==4 else False
X = linearize(X) 

params = {
    "rgb": rgb,
    "x_translate_max":10, 
    "y_translate_max":10, 
    "theta_range":(-np.pi / 4, np.pi / 4),
    "x_scale_range": (0.5, 2),
    "y_scale_range": (0.5, 2),
}

ds = InfiniteImageDataset(X, None, 50, params)
```

```python
ds.load()
grid_plot(get_2d_square_image_view(ds.X, rgb=rgb), imshow_options={"cmap": "gray"})
plt.show()
```
