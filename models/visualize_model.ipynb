```python
%load_ext autoreload

%autoreload

import pylearn2
from pylearn2.utils import serial

from pylearn2.datasets.fonts import Fonts
import theano
import numpy as np

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
model = serial.load("model.pkl")
data = Fonts(kind='all', accept_only='(.*)-[abcdxyz]-(.*)', labels_kind='letters')
#data = Fonts(kind='all')
labels = data.y
data = data.X.astype(np.float32)

```

```python
X = model.get_input_space().make_theano_batch()
X_code = model.encode(X)
X_hat = model.decode(X_code)


get_code = theano.function([X], X_code)
get_reconstruction = theano.function([X], X_hat)
decode = theano.function([X_code], X_hat)

```

```python
data_reconstruction = get_reconstruction(data)
data_code = get_code(data)
l = int(np.sqrt(data.shape[1]))
w, h = l, l
data_img = data.reshape( (data.shape[0], w, h) )
data_reconstruction_img= data_reconstruction.reshape( data_img.shape )
```

```python
# Generation as interpolation
def generate(nb_examples, data_code, sparsity=0.999):
    weights = np.random.uniform( size=(nb_examples, data_code.shape[0]) )
    masks = np.random.uniform( size = weights.shape ) <= (1 - sparsity)
    weights = weights * masks
    weights = weights / (weights.sum(axis=1)[:, np.newaxis])
    examples_codes = np.dot(weights, data_code).astype(np.float32)
    return decode(examples_codes)


nb = 32
per_dim = int(np.sqrt(nb))
generated = generate(nb, data_code)
generated_img = generated.reshape( (generated.shape[0], data_img.shape[1], data_img.shape[2] ) )
k = 1
for y in xrange(per_dim):
    for x in xrange(per_dim):
        plt.subplot(per_dim, per_dim, k)
        plt.axis('off')
        plt.imshow(generated_img[k - 1], cmap='gray')
        k += 1
plt.show()
```
