```python
from sklearn.neural_network import BernoulliRBM
import numpy as np
import matplotlib.pyplot as plt
from sampling import sample, dbn_sample
%matplotlib inline
import time

from representation import show_representation, get_representation
```

```python
hidden = (300,)

# build models
models = []
for layer in hidden:
    rbm = BernoulliRBM(n_components=layer, n_iter=1000, verbose=False, learning_rate=0.2)
    models.append(rbm)
```

```python
# load data
data = np.load("../ds/ds_b_GF_clean.npy")
data = data.reshape( (data.shape[0], np.prod(data.shape[1:]) ) )
```

```python
# train data
p = time.time()

D = data
for model in models:
    model.fit(D)
    D = model.transform(D)
print "took %.2fs to train" % (time.time() - p)
```

```python
# build  representations of layers

top = 500
imsize = (16, 16)

representations = []
cur_model_repr = get_representation(models[0], None, imsize=imsize)
representations.append(cur_model_repr)
for model in models[1:]:
    cur_model_repr = get_representation(model, cur_model_repr, top=top)
    representations.append(cur_model_repr)

```

```python
# show representations of layers

for r in representations:
    show_representation(r)
```

```python
# Generate samples
samples = []
samples = sample(model, steps=1000, nb_samples=80)
samples = np.array(samples)
samples = samples.reshape( (samples.shape[0], imsize[0], imsize[1]) )
show_representation(samples)
    
```
